```python
import pandas as pd

data = {
    "label": ["ham", "spam", "ham", "spam", "ham"],
    "message": [
        "Hey, how are you?",
        "WIN a free ticket now!!!",
        "Are we still meeting today?",
        "Congratulations, you won cash prize $$$",
        "See you at the party"
    ]
}

df = pd.DataFrame(data)
df.head()

df['clean_msg'] = df['message'].str.lower()
df['clean_msg'] = df['clean_msg'].str.replace(r'[^a-z0-9\s]', '', regex=True)
df[['message', 'clean_msg']].head()


df['tokens'] = df['clean_msg'].str.split()
df[['clean_msg', 'tokens']].head()

stopwords = {"the", "a", "is", "you", "to", "for", "in"}  # small custom set

df['tokens_no_stop'] = df['tokens'].apply(
    lambda words: [w for w in words if w not in stopwords]
)
df[['tokens', 'tokens_no_stop']].head()


from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

df['stems'] = df['tokens_no_stop'].apply(
    lambda words: [stemmer.stem(w) for w in words]
)
df[['tokens_no_stop', 'stems']].head()

df['word_count'] = df['tokens'].str.len()
df['digit_count'] = df['message'].str.count(r'\d')
df['exclamation_count'] = df['message'].str.count(r'!')
df[['message', 'word_count', 'digit_count', 'exclamation_count']].head()


```
